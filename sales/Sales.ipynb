```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Настройка визуализации
plt.style.use('ggplot')
%matplotlib inline
```

```python
sales_df = pd.read_csv('sales_data.csv')

# Проверка загрузки
sales_df.head()
```

```python
# Просмотр первых строк
sales_df.head()

# Информация о типах данных и пропусках
sales_df.info()

# Статистика по числовым колонкам
sales_df.describe()
```

```python
# Создание колонки 'Revenue'
sales_df['Revenue'] = sales_df['Quantity'] * sales_df['Price']

# Преобразование даты
sales_df['Date'] = pd.to_datetime(sales_df['Date'])
sales_df['Month'] = sales_df['Date'].dt.month
sales_df['DayOfWeek'] = sales_df['Date'].dt.dayofweek  # 0=понедельник
```

```python
sales_df.columns
```

```python
sales_df.head()
```

```python
# Выручка по категориям
revenue_by_category = sales_df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
revenue_by_category
```

```python
# Группировка по месяцам
monthly_sales = sales_df.groupby('Month')['Revenue'].sum()

# Визуализация
plt.figure(figsize=(10, 5))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Динамика продаж по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Выручка')
plt.xticks(rotation=0)
plt.show()
```

```python
# По количеству проданных единиц
top_products = sales_df.groupby('Product')['Quantity'].sum().nlargest(5)

# Визуализация
plt.figure(figsize=(10, 5))
top_products.plot(kind='barh', color='lightgreen')
plt.title('Топ-5 самых продаваемых товаров')
plt.xlabel('Количество проданных единиц')
plt.show()
```

```python
# Распределение выручки по категориям (круговая диаграмма)
plt.figure(figsize=(8, 8))
revenue_by_category.plot(kind='pie', autopct='%1.1f%%')
plt.title('Доля выручки по категориям')
plt.ylabel('')
plt.show()
```
